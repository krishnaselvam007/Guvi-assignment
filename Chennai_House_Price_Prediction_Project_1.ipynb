**KNN Modelling**

```python
from sklearn.neighbors import KNeighborsRegressor
from mlxtend.plotting import plot_decision_regions
from sklearn.model_selection import cross_val_score

def knn_no_plot(data, k, p):
    x = data[['AREA', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM', 'SALE_COND', 'PARK_FACIL', 'UTILITY_AVAIL', 'STREET', 'MZZONE', 'BUILD_AGE_YEAR', 'Commercial', 'House', 'Others']].values 
    # independent features # 'AREA', 'INT_SQFT', 'N_BEDROOM', 'N_BATHROOM', 'N_ROOM', 'SALE_COND', 'PARK_FACIL', 'UTILITY_AVAIL', 'STREET', 'MZZONE', 'BUILD_AGE_YEAR', 'Commercial', 'House', 'Others'
    y = data['SALES_PRICE'].values # y -> target value
    reg = KNeighborsRegressor(n_neighbors=k, p=p) #initialise the model with neighbours as k - hyper parameter p=1 - manhattan p=2 - Euclioden
    reg.fit(x, y) # train the model
    print("K : ", k, " Train Accuracy : ", reg.score(x,y), " Val Accuracy : ", 
          np.mean(cross_val_score(reg, x, y, cv=10)))  # test the model and it computes the accuracy (train data accuracy)
```

```python
for i in [1,2,3,4,5,20,30,40,80]:
  knn_no_plot(housedata,i, 1) # Manhattan Distance
```

```python
for i in [1,2,3,4,5,20,30,40,80]:
  knn_no_plot(housedata,i, 2) # Euclidean Distance
```

Therefore best k value is **20** in **Euclidean distance** and **5** in **manhattan distance**
